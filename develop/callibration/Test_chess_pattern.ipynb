```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
rows, columns, square_width = 6,9,27
```

```python
image = cv2.imread('E:/Jurij/Faks/MAGISTRSKA/GIT/3D_object_detection-w_web_camera/develop/callibration/images/frame_0.jpg')
```

```python
%matplotlib inline
def plot_image(image):
    f, axarr = plt.subplots(1,1, figsize = (8,8))
    axarr.axis('off')
    axarr.imshow(image, cmap='gray')
    axarr.plot()
```

```python
plot_image(image)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
ret, corners = cv2.findChessboardCorners(gray, (columns, rows), flags=cv2.CALIB_CB_ACCURACY)
```

```python
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 30, 0.001)

corners2 = cv2.cornerSubPix(gray, corners, (11, 11), (-1, -1), criteria)
```

```python
image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
cv2.drawChessboardCorners(image, (columns, rows), corners2, ret)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
#plot_image(image)
```
